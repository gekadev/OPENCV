```python
import cvzone
from cvzone.HandTrackingModule import HandDetector
from cvzone.ClassificationModule import Classifier
import cv2
import numpy as np
import time
import mediapipe 
from cvzone.FPS import FPS
import os
import math
import csv
import time

############# our Variables ############
wCam, hCam = 640, 480
cap = cv2.VideoCapture(0)
cap.set(3,wCam)
cap.set(4,hCam)
listImageModes   = []
listImgIcons     = []
selection = -1
counter   =0
modePositions = [(1136, 196), (1000, 384), (1136, 581)] # mode  to draw elips 
modeType = 0
selectionSpeed = 15
counterPause = 0
reset = 1
counterPause =0
selectionList = [-1, -1, -1,-1]

############################### 
detector = HandDetector(staticMode=False,maxHands=2,modelComplexity=1,detectionCon=0.5, minTrackCon=0.5)   

#import modes images
mainModePath ='Resources/Modes/'
for imageMode in os.listdir(mainModePath) :
    listImageModes.append(cv2.imread(os.path.join(mainModePath,imageMode)))

# importing all the icons to a list
folderPathIcons = "Resources/Icons"
listImgIconsPath = os.listdir(folderPathIcons)
for imgIconsPath in listImgIconsPath:
    listImgIcons.append(cv2.imread(os.path.join(folderPathIcons, imgIconsPath)))
#import background image     
imageBackground = cv2.imread('Resources/Background.png')
while True :
    _,frame = cap.read()
    frame = cv2.flip(frame,1)
    imageBackground [139:139+480,50:50+640]=frame
    imageBackground [0:720,847:1280]=listImageModes[modeType]
    #start detect hand
    hands , _ = detector.findHands(frame,draw=True,flipType=True)
    if hands and modeType <4 and counterPause ==0:
        hand = hands[0]
        #startt detect whic finger is up
        if detector.fingersUp(hand)==[0,1,0,0,0]:
            if selection != 1:
                counter=1
            selection = 1
        elif detector.fingersUp(hand)==[0,1,1,0,0]:
            if selection != 2:
                counter=1
            selection = 2
        elif detector.fingersUp(hand)==[0,1,1,1,0]:
            if selection != 3:
                counter=1
            selection = 3 
        else :
            selection =-1 
            counter   = 0

        if counter >0 :
            counter +=1
            if modeType !=3:
              cv2.ellipse(imageBackground,modePositions[selection-1],(103,103),0,0,counter*selectionSpeed,(0,255,0),20) 
    #1- start detect hand   
        if counter * selectionSpeed >360 :
            selectionList[modeType]=selection
            modeType +=1
            counter = 0
            selection = -1
            counterPause =1

     #reste index to main menu after order placed image  
        if modeType == 4 :
                modeType = 0
                imageBackground [0:720,847:1280]=listImageModes[modeType]
                selection =-1 
                counter   = 0 
     
        else:
            #add icon based on selection  in case mode not eual 4
            if selectionList[0] != -1:
                imageBackground[636:636 + 65, 133:133 + 65] = listImgIcons[selectionList[0] - 1]
            if selectionList[1] != -1:
                imageBackground[636:636 + 65, 340:340 + 65] = listImgIcons[2+selectionList[1]]
            if selectionList[2] != -1:
                imageBackground[636:636 + 65, 542:542 + 65] = listImgIcons[5+selectionList[2]]          

                     
        #pause counting after each selection
    if counterPause >0:
            counterPause +=1
            if counterPause >10 :
                counterPause =0  

                
#reset to main menu if  no hands need to reset every think
    elif not hands and modeType <4 and counterPause == 0 :
                modeType = 0
                imageBackground [0:720,847:1280]=listImageModes[modeType]
                selection =-1 
                counter   = 0
                counterPause =0
                selectionList=[-1,-1,-1,-1]
                '''
                if selectionList[0] == -1:
                    imageBackground[636:636 + 65, 133:133 + 65] = (0,0,3)
                if selectionList[1] == -1:
                    imageBackground[636:636 + 65, 340:340 + 65] = (0,0,3)
                if selectionList[2] == -1:
                    imageBackground[636:636 + 65, 542:542 + 65] = (0,0,3)          
            '''
                     
    #cv2.imshow('White Image ',imageWhite)
    imgPNG = cv2.imread("C:/Users/Geka/Desktop/update.png",cv2.IMREAD_UNCHANGED)
    imgOverlay = cvzone.overlayPNG(imageBackground, imgPNG, pos=[50, 566])
    #cv2.imshow('Main Image ',frame)
    cv2.imshow('Order Now  ',imageBackground)


    # Display the camera feed
    key = cv2.waitKey(10)
    if key == ord('q'):
        break
    
cap.release()
cv2.destroyAllWindows() 
```

```python
import cv2
import numpy as np
import time

# Initialize variables
width, height = 400, 300
start_point = (50, 50)
end_point = (350, 250)
line_thickness = 2
num_steps = 100  # Adjust the number of steps
delay_between_steps = 0.01  # Adjust the delay between steps (in seconds)

# Create a black image
image = np.zeros((height, width, 3), dtype=np.uint8)

# Calculate the step size for each coordinate
step_x = (end_point[0] - start_point[0]) / num_steps
step_y = (end_point[1] - start_point[1]) / num_steps

# Main loop
for step in range(num_steps + 1):
    # Create a black image for each frame
    frame = np.zeros((height, width, 3), dtype=np.uint8)

    # Calculate the current position of the line
    current_x = int(start_point[0] + step * step_x)
    current_y = int(start_point[1] + step * step_y)
    if current_x >300 or current_y >200 :
        current_x = 350
        current_y =250

    # Draw the line up to the current position
    cv2.line(frame, start_point, (current_x, current_y), (0, 0, 255), line_thickness)

    # Display the frame
    cv2.imshow('Drawing Line', frame)

    # Pause for a moment before moving to the next frame
    time.sleep(delay_between_steps)

    # Break the loop if 'q' key is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture object and close the window
cv2.destroyAllWindows()
```
