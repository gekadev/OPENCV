# Drowsy Driver

```python
import pandas as pd
import numpy as np
from scipy.spatial import distance
from imutils import face_utils
import imutils
import dlib
import cv2
import datetime
import playsound
import time


#strat load face detection 

face_detection =  dlib.get_frontal_face_detector()
predictor  = dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')
# get face points 
(lstart,lend) = face_utils.FACIAL_LANDMARKS_68_IDXS['left_eye']
(rstart,rend) = face_utils.FACIAL_LANDMARKS_68_IDXS['right_eye']
#### validate parameters ##############
videoName = 'accident.mp3'
flag = 0
frame_check = 20
thresh = .23
# claculate asp reation
def calc_asp_ratio (eye):
    x = distance.euclidean(eye[1],eye[5])
    y = distance.euclidean(eye[2],eye[4])
    c = distance.euclidean(eye[0],eye[3])
    ear = (x+y)/(c*2)
    return ear

cap = cv2.VideoCapture(0)
while True :
    _ ,frame = cap.read()
    frame = imutils.resize(frame,height=640,width=640)
    grayImage = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    # start detecting face
    faces = face_detection(grayImage)
    for face in faces :
        #start detect points
        x1 = face.left()
        y1 = face.top()
        x2 = face.right()
        y2 = face.bottom() 
        cv2.rectangle(frame,(x1,y1),(x2,y2),(0,255,0),2)
        #start to get landmarks
        landmarks = predictor(grayImage,face)
        landmarks = face_utils.shape_to_np(landmarks)
        #getspecific land marks for eyes 
        leftEye = landmarks[lstart:lend]
        rightEye =landmarks[rstart:rend]
        #claculate aspreatio
        lratio = calc_asp_ratio(leftEye)
        rratio = calc_asp_ratio(rightEye)
        asp_ratio = (lratio+rratio)/2
        lHull= cv2.convexHull(leftEye)
        rHull = cv2.convexHull(rightEye)
        #draw contour
        cv2.drawContours(frame,[lHull],-1,(0,255,0),2)
        cv2.drawContours(frame,[rHull],-1,(0,255,0),2)
        # print(asp_ratio)
        if asp_ratio < thresh :
            flag +=1
            if flag >=frame_check :
                    time.sleep(0.01)
                    cv2.putText(frame, "Driver Is Sleaping ", (10, 30),cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)
                    playsound.playsound(videoName, True)                      
        else :
            falg = 0
            cv2.putText(frame, "Normal ", (10, 30),cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)
            continue            

    cv2.imshow('image',frame)
    key = cv2.waitKey(10)
    if key == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()  
```

# Drowsy Driver Using OOP

```python
from scipy.spatial import distance
from imutils import face_utils
import imutils
import dlib
import cv2
import datetime
import playsound
import time


class Ifacedetection():
    def detectFace(self):
        pass
    
    def detectLandmarks(self):
        pass
    
    def detectSpecificPart(self):
        pass
    def convexHull(self):
        pass
    
    def drawContour(self):
        pass
    
    
class Icalculation():
    
    def calc_asp_ratio (self):
        pass
    
    def firist_distance(self):
        pass 
    
    def second_distance(self):
        pass 
    
    def third_distance(self):
        pass 


    
class facedetection(Ifacedetection):
    
    def __init__(self):
        pass

    def detectFace(self):
        return dlib.get_frontal_face_detector()
    
    def detectLandmarks(self):
        return dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')

    
    def detectSpecificPart(self,part):
        (start,end) = face_utils.FACIAL_LANDMARKS_68_IDXS[part]
        return (start,end)
    
    def convexHull(self,landMarksPoints):
        return cv2.convexHull(landMarksPoints)
        
    def drawContour(self,hull):   
        return cv2.drawContours(frame,[hull],-1,(0,255,0),1)

    
    
class calculation(Icalculation):
    def __init__(self,part,face_object:Ifacedetection):
        self.__part = part
        self.__face_object = face_object

    def getpart(self):
        return self.__part    

    
    def firist_distance(self):
        x =distance.euclidean(self.__part[1],self.__part[5])
        return x
    
    def second_distance(self):
        y = distance.euclidean(self.__part[2],self.__part[4])
        return y 
    
    def third_distance(self):
        z = distance.euclidean(self.__part[0],self.__part[3])
        return z     
    
    def calc_asp_ratio (self):
        ratio = (self.firist_distance()+self.second_distance())/(self.third_distance()*2)
        return ratio

    

class IImagePreprocessing():
    def resizeImage(self):
        pass
    
    def convertToGray(self):
        pass

    
    
class ImagePreprocessing(IImagePreprocessing):

    def __init__(self,image):
        self.__image = image
        
       
        
    def resizeImage(self):
        return imutils.resize(self.__image,width=640,height=640)
    

    def convertToGray(self):
        return cv2.cvtColor(self.__image,cv2.COLOR_BGR2GRAY)
  
    
    def run_preprocessing(self):
        image     = self.resizeImage()
        grayImage = self.convertToGray()
        return grayImage  
```

```python
#create face object

face_object = facedetection()
face_detection = face_object.detectFace()
predictor = face_object.detectLandmarks()
(lstart,lend) = face_object.detectSpecificPart('left_eye')
(rstart,rend) = face_object.detectSpecificPart('right_eye')
videoName = 'accident.mp3'
flag = 0
frame_check = 20
thresh = .23
cap = cv2.VideoCapture(0)
while True :
    _,frame = cap.read()
    pre =  ImagePreprocessing(frame) 
    grayImage = pre.run_preprocessing()
    #start detect face
    faces =face_detection(grayImage) 
    for face in faces :
        x1,y1,x2,y2= face.left(),face.top(),face.right(),face.bottom()
        cv2.rectangle(frame,(x1,y1),(x2,y2),(0,255,0),3)
        landmark = predictor(grayImage,face)
        landmark = face_utils.shape_to_np(landmark)
        leftEye  = landmark[lstart:lend]
        rightEye = landmark[rstart:rend]
        calc_l   = calculation(leftEye,face_object)
        calc_r   = calculation(rightEye,face_object)
        leftEyeRatio  = calc_l.calc_asp_ratio()
        rightEyeRatio = calc_r.calc_asp_ratio()
        asp_ratio     = (leftEyeRatio+rightEyeRatio)/2
        leftEyeHull   = face_object.convexHull(leftEye)
        rightEyeHull  = face_object.convexHull(rightEye)
        face_object.drawContour(leftEyeHull)
        face_object.drawContour(rightEyeHull) 
        # print(asp_ratio)
        if asp_ratio < thresh :
            flag +=1
            if flag >=frame_check :
                    time.sleep(0.01)
                    cv2.putText(frame, "Driver Is Sleaping ", (10, 30),cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)
                    playsound.playsound(videoName, True)                      
        else :
            falg = 0
            cv2.putText(frame, "Normal ", (10, 30),cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)
                        
        cv2.imshow('frame',frame)
 
    key = cv2.waitKey(10)
    if key == ord('q'):
        break
cap.release()
cv2.destroyAllWindows() 
```
