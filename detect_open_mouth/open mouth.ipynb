# Detect Open mouth 

```python
from scipy.spatial import distance
from imutils import face_utils
import imutils
import dlib
import cv2
import datetime
import playsound
import time


#define face detection and lanmarks part 

face_detection  = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')
# define parts of lanmarks
(startface,endface) = face_utils.FACIAL_LANDMARKS_68_IDXS['mouth']


def caluslateMAR(mouth):
    A = distance.euclidean(mouth[2],mouth[10]) # 51 -59
    B = distance.euclidean(mouth[4],mouth[8])  #53,57
    C = distance.euclidean(mouth[0],mouth[6])  #0,6
    mar = (A+B)/(C*2)
    return mar

thresh = .6
cap = cv2.VideoCapture(0)
while True :
    _,frame = cap.read()
    frame = imutils.resize(frame,height=640,width=640)
    grayImage =cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces = face_detection(grayImage)
    #start face detection
    for face in faces :
        x1 = face.left()
        y1 = face.top()
        x2 = face.right()
        y2 = face.bottom()
        cv2.rectangle(frame,(x1,y1),(x2,y2),(255,0,0),3)
         # start define land marks of object   
        landmarks = predictor(grayImage,face)
        landmarks =face_utils.shape_to_np(landmarks)
        #get mouth klandmarks
        mouth = landmarks[startface:endface]
        #calculaet MAR Ratio
        ratio = caluslateMAR(mouth)
        mouthhull = cv2.convexHull(mouth)
        #start draw convhyul
        cv2.drawContours(frame,[mouthhull],-1,(255,0,0),3)
        #strat predction
        # print(ratio)
        if ratio >thresh :
            time.sleep(0.01)
            cv2.putText(frame, 'Mouth Is Open' , (10, 30),cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 0, 0), 2)
            
    cv2.imshow('image',frame)
    key = cv2.waitKey(10)
    if key == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()  
```

# Detect Open Mouse Using OPP

```python
from scipy.spatial import distance
from imutils import face_utils
import imutils
import dlib
import cv2
import datetime
import playsound
import time


class Ifacedetection():
    def detectFace(self):
        pass
    
    def detectLandmarks(self):
        pass
    
    def detectSpecificPart(self):
        pass
    def convexHull(self):
        pass
    
    def drawContour(self):
        pass
    
    
class Icalculation():
    
    def calc_asp_ratio (self):
        pass
    
    def firist_distance(self):
        pass 
    
    def second_distance(self):
        pass 
    
    def third_distance(self):
        pass 


    
class facedetection(Ifacedetection):
    
    def __init__(self):
        pass

    def detectFace(self):
        return dlib.get_frontal_face_detector()
    
    def detectLandmarks(self):
        return dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')

    
    def detectSpecificPart(self,part):
        (start,end) = face_utils.FACIAL_LANDMARKS_68_IDXS[part]
        return (start,end)
    
    def convexHull(self,landMarksPoints):
        return cv2.convexHull(landMarksPoints)
        
    def drawContour(self,hull):   
        return cv2.drawContours(frame,[hull],-1,(0,255,0),1)

    
    
class calculation(Icalculation):
    def __init__(self,part,face_object:Ifacedetection):
        self.__part = part
        self.__face_object = face_object

    def getpart(self):
        return self.__part        

    
    def firist_distance(self):
        x =distance.euclidean(self.__part[2],self.__part[10])
        return x
    
    def second_distance(self):
        y = distance.euclidean(self.__part[4],self.__part[8])
        return y 
    
    def third_distance(self):
        z = distance.euclidean(self.__part[0],self.__part[6])
        return z     
    
    def calc_asp_ratio (self):
        ratio = (self.firist_distance()+self.second_distance())/(self.third_distance()*2)
        return ratio

    

class IImagePreprocessing():
    def resizeImage(self):
        pass
    
    def convertToGray(self):
        pass

    
    
class ImagePreprocessing(IImagePreprocessing):

    def __init__(self,image):
        self.__image = image
        
       
        
    def resizeImage(self):
        return imutils.resize(self.__image,width=640,height=640)
    

    def convertToGray(self):
        return cv2.cvtColor(self.__image,cv2.COLOR_BGR2GRAY)
  
    
    def run_preprocessing(self):
        image     = self.resizeImage()
        grayImage = self.convertToGray()
        return grayImage  
```

```python
#create face object

face_object = facedetection()
face_detection = face_object.detectFace()
predictor = face_object.detectLandmarks()
(start,end) = face_object.detectSpecificPart('mouth')
thresh = .6
cap = cv2.VideoCapture(0)
while True :
    _,frame = cap.read()
    pre =  ImagePreprocessing(frame) 
    grayImage = pre.run_preprocessing()
    #start detect face
    faces =face_detection(grayImage) 
    for face in faces :
        x1,y1,x2,y2= face.left(),face.top(),face.right(),face.bottom()
        cv2.rectangle(frame,(x1,y1),(x2,y2),(0,255,0),3)
        landmark = predictor(grayImage,face)
        landmark = face_utils.shape_to_np(landmark)
        mouth = landmark[start:end]
        calc  = calculation(mouth,face_object)
        asp_ratio      = calc.calc_asp_ratio()
        mouthHull  = face_object.convexHull(mouth)
        face_object.drawContour(mouthHull)
        # print(asp_ratio)
        if asp_ratio > thresh :  
            time.sleep(0.01)
            cv2.putText(frame, 'Mouth Is Open' , (10, 30),cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 0, 0), 2)
                        
        cv2.imshow('frame',frame)
 
    key = cv2.waitKey(10)
    if key == ord('q'):
        break
cap.release()
cv2.destroyAllWindows() 
```
